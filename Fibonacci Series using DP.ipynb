Fibonacci series using Recursion

```python
# method implementation using recursion
def fib(n):
  #base case condition
  if n == 0 or n == 1:
    return n
  else:
    result = fib(n-1) + fib(n-2)
  return result

# Driver code
print(fib(7))
```

Fibonacci series using Dynamic Programming

**1. Memoization (Top-Down Approach)**

TC - O(n) SC - O(n)

```python
# Memoization implementation
def fib_topDown(n, memo):
  if memo[n] is not None:
    return memo[n]
  # Base case
  if n == 0 or n == 1:
    return n
  else:
    # Recursion
    result = fib_topDown(n -1, memo) + fib_topDown(n-2, memo)

  # Storing the value
  memo[n] = result
  return result


def fib_memo(n):
  memo = [None]*(n + 1)
  # Function call
  result = fib_topDown(n, memo)
  return result

# Driver code
print(fib_memo(50))
```

**2. Tabulation (Bottom-Up Approach)**

TC-O(n) SC-O(n)

```python
# implementation of tabulation approach
def fib_bottomUp(n):
  bottomUp = [None]*(n+1)
  bottomUp[0] = 0
  bottomUp[1] = 1
  # No recursive call
  for i in range(2, n+1):
    bottomUp[i] = bottomUp[i-1] + bottomUp[i-2]
  return bottomUp[n]

# Driver code
print(fib_bottomUp(1000))
```

Highly optimized approach

Tabulation where SC is O(1)

```python
# Tabulation implementation without storing results
def fib_bottomUp(n):
  first = 0
  second = 1
  for i in range(2, n+1):
    next = first + second
    first, second = second, next
  return next

# Driver code
print(fib_bottomUp(10000))
```
