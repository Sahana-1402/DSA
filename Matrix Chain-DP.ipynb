Recursive brute force approach

```python
import sys

def matrixChainMul(p, i, j):
  #base case
  if i == j:
    return 0

  minVal = sys.maxsize

  for k in range(i, j):
    count = matrixChainMul(p, i, k) + matrixChainMul(p, k+1, j) + p[i-1] * p[k] * p[j]

    if count < minVal:
      minVal = count

  return minVal

# Driver code
arr = [1, 3, 1, 2, 3]
N = len(arr)

matrixChainMul(arr, 1, N-1)
```

Memoization

```python
def matrixChainMul(p, i, j):
  dp = [[-1]*100 for i in range(100)]

  if i == j:
    return 0

  if dp[i][j] != -1:
    return dp[i][j]

  dp[i][j] = sys.maxsize
  for k in range(i, j):
    dp[i][j] = matrixChainMul(p, i, k) + matrixChainMul(p, k+1, j) + p[i-1]*p[k]*p[j]

  return dp[i][j]


# Driver code
arr = [1, 3, 1, 2, 3]
N = len(arr)

matrixChainMul(arr, 1, N-1)
```
