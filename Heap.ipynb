Implement heap data structure

```python
def buildHeap(arr, n):
  startIndex = n//2 - 1
  for i in range(startIndex, -1, -1):
    heapify(arr, n, i)
  return arr

def heapify(arr, n, i):
  smallest = i
  l = 2*i + 1
  r = 2*i + 2
  print(l)
  if l<n and r<n:
    if arr[l] < arr[smallest]:
      smallest = l
    elif arr[r] < arr[smallest]:
      smallest = r

    heapify(arr, n, smallest)

#Driver
arr = [0,3,7,5,1,9,6]
n = 7
print(buildHeap(arr,n))
```
