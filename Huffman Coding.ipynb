Huffman coding Algortihm Implementation

```python
import heapq

class Node:
  def __init__(self, freq, char, left = None, right = None):
    self.freq = freq
    self.char = char
    self.left = left
    self.right = right
    self.huff = ''

  def __lt__(self, other):
        return self.freq < other.freq

def printNodes(node, val=''):
    newVal = val + str(node.huff)

    if(node.left):
        printNodes(node.left, newVal)
    if(node.right):
        printNodes(node.right, newVal)

    if(node.left is None and node.right is None):
        print(f"{node.char} - {newVal}")

def hufffmanCoding(chars):
  pq = []
  for char in chars:
    new_node = Node(chars[char], char)
    heapq.heappush(pq, new_node)

  while len(pq) > 1:
    left = heapq.heappop(pq)
    right = heapq.heappop(pq)

    left.huff = '0'
    right.huff = '1'

    heapq.heappush(pq, Node(left.freq+right.freq, left.char+right.char, left, right))

  printNodes(pq[0])

# Driver code
# Characters and freq given
chars = {'a':5, 'b':9, 'c':12, 'd':13, 'e':16, 'f':45}

hufffmanCoding(chars)
```
