Job Sequencing with deadline

Constraint: Every job needs to be completed on or before the max deadline

Optimisation: Max profit

```python
# method definition
# Time Complexity- O(n^2) Space complexity-O(n)
def jobScheduling(arr, maxDeadline):
  n = len(arr)
  arr.sort(key = lambda x:x[1], reverse = True)

  result = [False]*maxDeadline
  job = [-1]*maxDeadline

  for i in range(n):
    for j in range(min(maxDeadline-1, arr[i][2]-1), -1, -1):
      if result[j] is False:
        result[j] = True
        job[j] = arr[i][0]
        break

  print(job)

# Driver code

# Job,Profit,Deadline
arr = [
    ['J1',55,5],
    ['J2',65,2],
    ['J3',75,7],
    ['J4',60,3],
    ['J5',70,2],
    ['J6',50,1],
    ['J8',68,5],
    ['J9',45,3]
]

jobScheduling(arr,7)
```
