Linked List - Creation, Insertion(at beginning, at end, in-between)

```python
class Node:
  def __init__(self, node_data):
    self.data = node_data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  # Insert after a node
  def insertAfterNode(self, prev_node, new_data):
    if prev_node is None:
      print("Previous node not present")
      return

    new_node = Node(new_data)
    new_node.next = prev_node.next
    prev_node.next = new_node

  # Insert a new node at the end
  def insertAtEnd(self, new_data):
    new_node = Node(new_data)

    if self.head is None:
      self.head = new_node
      return

    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  # Method to enter new nod at the begining
  def insertAtBeginning(self, new_data):
    new_node = Node(new_data)
    new_node.next = self.head
    self.head = new_node

  # Traverse the node and print the node data
  def printNodeData(self):
    temp = self.head
    while temp:
      print(str(temp.data) + " ", end=" ")
      temp = temp.next

# Driver Code
llist = LinkedList()
llist.insertAtBeginning(10)
llist.insertAtBeginning(20)
llist.insertAtBeginning(30)
llist.insertAtEnd(5)
llist.insertAfterNode(llist.head.next,15)
llist.insertAfterNode(llist.head,25)

llist.printNodeData()
```
