Implementation of fractional Knapsack Problem using Greedy Algorithm

```python
# Item class with profit and weight
class Item:
  def __init__(self, profit, weight):
    self.profit = profit
    self.weight = weight

# Method implementation of the knapsack using greedy algorithm
def fractionalKnapsack(arr, M):
  # sort the arr based on profit/weight
  arr.sort(key = lambda x:(x.profit/x.weight), reverse=True)
  for vals in arr:
    print(vals.profit, vals.weight, (vals.profit/vals.weight))

  max_profit = 0.0
  for items in arr:
    if items.weight <= M:
      M -= items.weight
      max_profit += items.profit
    else:
      max_profit += items.profit * M/items.weight
      break

  return max_profit


# Driver code
M = 37
arr = [Item(25, 5), Item(75, 10), Item(100, 12), Item(50, 4), Item(45, 7), Item(90, 9), Item(30, 3)]

maxProfit = fractionalKnapsack(arr, M)
print("The maximum profit is", maxProfit)
```
