Non-Comparision sorting algorithm - Count Sort

```python
def count_sort(arr):
  max_val =  max(arr)
  count_arr = [0]*(max_val+1)
  for i in range(len(arr)):
    count_arr[arr[i]] += 1

  cumm_arr = [0]*(max_val+1)
  for i in range(1, max_val+1):
    cumm_arr[i] = count_arr[i] + cumm_arr[i-1]

  sorted_arr = [0]*len(arr)
  # from last- to return in-place sorted array
  for i in range(len(arr)-1, -1, -1):
    pos = cumm_arr[arr[i]]
    cumm_arr[arr[i]] -= 1
    sorted_arr[pos-1] = arr[i]

  return sorted_arr

arr = [1,4,1,2,7,5,2]
print(count_sort(arr))
```
