Implementation of Dijkstra's Algortithm

```python
import heapq

# Method defination
def calculateDistances(graph, source):
  vertices = len(graph)
  distances = {vertex: float('infinity') for vertex in graph}
  distances[source] = 0

  pq = [(0,source)]
  while pq:
    current_distance, current_vertex = heapq.heappop(pq)

    if current_distance > distances[current_vertex]:
      continue

    for neighbour, distance in graph[current_vertex].items():
      total_distance = current_distance + distance
      if total_distance < distances[neighbour]:
        distances[neighbour] = total_distance
        heapq.heappush(pq, (total_distance,neighbour))

  return distances

# Driver code
# Graph as an Adjacency list
graph = {
    'A':{'B':2, 'C':5, 'D':2, 'E':7, 'F':50},
    'B':{'C':2, 'D':1, 'E':2, 'F':60},
    'C':{'B':3, 'E':2, 'F':90},
    'D':{'E':1, 'F':3},
    'E':{'D':4, 'F':4},
    'F':{}
}

print("The shortest path from a fixed source A is")
print(calculateDistances(graph,'A'))
```
