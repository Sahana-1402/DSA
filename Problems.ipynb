1. Kth smallest element

    (Solved using heaps)

```python
import heapq

# method defination
def KsmallestElement(arr, k):
  # O(n)
  heapq.heapify(arr)
  # O(logn)
  result = heapq.nsmallest(k, arr)
  return result[k-1]

# Driver code
arr = [40, 25, 68, 79, 52, 66, 89, 97]
k = 4
result = KsmallestElement(arr, k)
print(result)
```

1. Kth smallest element

    (Solved using selection procedure)

```python
# method implementation of partition
# total time complexity- Best/Avg = O(n), Worst = O(n^2)
# TC for partition - O(n)
def partition(arr, lower, upper):
  pivot = arr[lower]
  i = lower
  for j in range(i+1, upper+1):
    if arr[j] <= pivot:
      i+=1
      arr[i], arr[j] = arr[j], arr[i]
  arr[i], arr[lower] = arr[lower], arr[i]
  return i

# method implementation
def KsmallestElement(arr, k, lower, upper):
  if lower == upper:
    return arr[lower]
  if lower < upper:
    pivot = partition(arr, lower, upper)
    requiredIndex = k-1
    if pivot == requiredIndex:
      return arr[requiredIndex]
    elif pivot < requiredIndex:
      return KsmallestElement(arr, k, pivot+1, upper)
    else:
      return KsmallestElement(arr, k, lower, pivot-1)


# Driver code
arr = [40, 25, 68, 79, 52, 66, 89, 97]
k = 2
result = KsmallestElement(arr, k, 0, len(arr)-1)
print(result)
```

2. Sort Colors

```python
# TC = O(n) SC = 0(1)
# Method implementation
def sortColors(nums):
  n = len(nums)
  cnt0, cnt1, cnt2 = 0, 0, 0
  for i in range(n):
    if nums[i] == 0:
      cnt0 += 1
    elif nums[i] == 1:
      cnt1 += 1
    else:
      cnt2 += 1
  nums = cnt0 * [0] + cnt1 * [1] + cnt2 * [2]
  print(nums)
  return nums

# Driver Code
nums = [2, 0, 2, 1, 1, 0, 2]
result = sortColors(nums)
print(result)
```

2. Sort Colors

    (using two pointer)

```python
# TC = O(n) SC = 0(1)
# Method implementation
def sortColors(nums):
  p, i = 0, 0
  q = len(nums) - 1
  while i < q:
    if nums[i] == 0:
      nums[p], nums[i] = nums[i], nums[p]
      i += 1
      p += 1
    if nums[i] == 2:
      nums[q], nums[i] = nums[i], nums[q]
      # [1, 0, 2]
      q -= 1
    else:
      i += 1
  return nums

# Driver Code
nums = [2, 0, 1]
result = sortColors(nums)
print(result)
```

3. Kth largest element

    (Solved using selection procedure)

```python
# method implementation of partition
# TC - O(n)
def partition(arr, lower, upper):
  pivot = arr[lower]
  i = lower
  for j in range(i+1, upper+1):
    if arr[j] <= pivot:
      i+=1
      arr[i], arr[j] = arr[j], arr[i]
  arr[i], arr[lower] = arr[lower], arr[i]
  return i

# method implementation
def KlargestElement(arr, k, lower, upper):
  if lower == upper:
    return arr[lower]
  if lower < upper:
    pivot = partition(arr, lower, upper)
    requiredIndex = len(arr) - k
    if pivot == requiredIndex:
      return arr[requiredIndex]
    elif pivot < requiredIndex:
      return KlargestElement(arr, k, pivot+1, upper)
    else:
      return KlargestElement(arr, k, lower, pivot-1)


# Driver code
arr = [40, 25, 68, 79, 52, 66, 89, 97]
k = 2
result = KsmallestElement(arr, k, 0, len(arr)-1)
print(result)
```

4. Majority element

    (Using Dict)

```python
from collections import Counter

# TC = O(n) SC = O(n)
# Method implementation
def majorityElement(nums):
  counts = Counter(nums)
  return max(counts.keys(), key=counts.get)

# Driver code
nums = [2, 2, 1, 1, 1, 2, 2]
result = majorityElement(nums)
print(result)
```

5. Find of peak element

    (Brute force)

```python
# Brute force approach TC = O(n)
# Method implementation
def peak(nums):
  n = len(nums)
  for i in range(1, n-1):
    if(nums[i-1] < nums[i] and nums[i] > nums[i+1]):
      return i
  return "No peak element"

# Driver code
nums = [1, 2, 3, 1]
result = peak(nums)
print(result)
```

5. Find of peak element

    (Modified binary search)

```python
# TC - O(logn)
# Method definition
def findPeak(nums):
  start, end = 0, len(nums) - 1

  if nums[0] > nums[1]:
    return 0

  if nums[-1] > nums[-2]:
    return len(nums) - 1

  while start < end:
    mid = start + (end-start) // 2
    if nums[mid] > nums[mid+1]:
      end = mid
    else:
      start = mid+1
  return end

# Driver code
nums = [1, 2, 3, 1]
result = findPeak(nums)
print(result)
```

6. Median of two sorted arrays

```python
# O(NlogN), N= m+n
def median(num1, num2):
  nums = sorted(num1 + num2)
  N = len(nums)
  if N%2 == 0:
    mid = N//2
    return (nums[mid]+nums[mid-1])/2
  else:
    return nums[N/2]

# Driver code
num1 = [1,3]
num2 = [2,4]
result = median(num1, num2)
print(result)
```

7. Power of x,n

```python
# Method defination
def findPower(x, n):
  if n == 0:
    return 1
  elif n < 0:
    x = 1/x
    n = -n
    return findPower(x, n)
  elif n == 1:
    return x
  else:
    mid = n//2
    b = findPower(x, mid)
    pow = b * b

    if n % 2 == 0:
      return pow
    else:
      return pow*x

# Driver code
x = 2
n = 4
result = findPower(x,n)
print(result)
```

8. Divide two integers (Brute force)

```python
# method defination
def divide(divident, divisor):
  cnt = 0
  while divident >= divisor:
    divident -= divisor
    cnt += 1
  return (cnt, divident)

# Driver code
result = divide(13,3)
print(result)
```
