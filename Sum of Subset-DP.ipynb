Given a set of non-nagtive integers, and a value sum, determine if there is a subset of the given set with a sum equal o the given sum value

Brute Force approach

```python
def isSubset(setInput, n, sum):
  # base case
  if sum == 0:
    return True
  if n == 0:
    return False

  if setInput[n-1] > sum:
    return isSubset(setInput,n-1, sum)

  # exclude the element or include it
  else:
    return isSubset(setInput, n-1, sum) or isSubset(setInput, n-1, sum - setInput[n-1])



# Driver code
given_set = [3, 24, 4, 12, 5, 2]
sum = 9
n = len(given_set)

print(isSubset(given_set, n, sum))
```

Memoization

```python
def isSubset(given_set, n, sum):
  K = [[None]*(sum+1) for i in range(n+1)]

  if sum == 0:
    return True
  if n == 0:
    return False
  if K[n][sum] is not None:
    return K[n][sum]
  if given_set[n-1] > sum:
    K[n][sum] = isSubset(given_set, n-1, sum)
    return K[n][sum]
  else:
    K[n][sum] = isSubset(given_set, n-1, sum) or isSubset(given_set, n-1, sum - given_set[n-1])
    return K[n][sum]

# Driver code
given_set = [3, 24, 4, 12, 8, 2]
sum = 9
n = len(given_set)

print(isSubset(given_set, n, sum))
```

Tabulation

```python
from re import T
def isSubset(given_set, n, sum):
  K = [[False]*(sum+1) for i in range(n+1)]

  for i in range(1, sum+1):
    K[0][i] = False
  for i in range(n+1):
    K[i][0] = True

  for i in range(1, n+1):
    for j in range(1, sum+1):
      if j < given_set[i-1]:
        K[i][j] = K[i-1][j]
      else:
        K[i][j] = K[i-1][j] or K[i-1][j-given_set[i-1]]

  return K[n][sum]

# Driver code
given_set = [3, 24, 4, 12, 8, 2]
sum = 9
n = len(given_set)

print(isSubset(given_set, n, sum))
```
