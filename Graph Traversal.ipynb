**Depth First Traversal - DFT**

```python
# Implementation of the depth first traversal
def DFT(v, visited, adjacency_list):
    if v not in visited:
      visited.add(v)
      print(v, end = ' ')
      for w in adjacency_list[v]:
        # Recursion call
        DFT(w, visited, adjacency_list)

# Adjacency list of the graph
adjacency_list = {
     '5':['3','7'],
     '3':['2','4'],
     '7':['8'],
     '2':[],
     '4':['8'],
     '8':[]
}

visited = set()

# Driver code
DFT('5', visited, adjacency_list)
```

**Breadth First Traversal - BFT**

```python
from collections import deque

queue = deque()

#Implementation of the breadth first traversal
def BFT(root, visited, queue):
  visited.add(root)
  queue.append(root)
  # When elements are there in the queue
  while queue:
    x = queue.popleft()
    print(x, end = ' ')

    for w in adjecency_list[x]:
      if w not in visited:
        visited.add(w)
        queue.append(w)

# Driver code
adjecency_list = {
    'A':['B','C','D'],
    'B':['E'],
    'C':['E','F'],
    'D':['F'],
    'E':['G'],
    'F':['G'],
    'G':[]
}

visited = set()

BFT('A', visited, queue)
```
