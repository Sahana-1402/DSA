Create a HashTable, with **Division Modulo hash function**

```python
class HashTable:
  def __init__(self):
    self.MAX = 10
    self.arr = [None for i in range(self.MAX)]

  def hash_function(self,key):
    hash = 0
    m = self.MAX
    for c in key:
      hash += ord(c)
    return hash % m

  def addItem(self, key, value):
    h = self.hash_function(key)
    self.arr[h] = value

  def getItem(self, key):
    h = self.hash_function(key)
    return self.arr[h]
```

```python
obj1 = HashTable()
```

```python
obj1.addItem("march 2021",23)
obj1.addItem("march 2025",28)
```

```python
obj1.getItem("march 2021")
```

The same HashTable that overrides in-built setitem and getitem methods

```python
class HashTable:
  def __init__(self):
    self.MAX = 10
    self.arr = [None for i in range(self.MAX)]

  def hash_function(self,key):
    hash = 0
    m = self.MAX
    for c in key:
      hash += ord(c)
    return hash % m

  def __setitem__(self, key, value):
    h = self.hash_function(key)
    self.arr[h] = value

  def __getitem__(self, key):
    h = self.hash_function(key)
    return self.arr[h]
```

```python
obj2 = HashTable()
```

```python
obj2["march 2021"] = 23
obj2["march 2025"] = 27
```

```python
obj2["march 2021"]
```

**COLLISION RESOLUTION TECHNIQUES**

1**. Chaining**
(with a list)

```python
class HashTable:
  def __init__(self):
    self.MAX = 10
    self.arr = [[] for i in range(self.MAX)]

  def hash_function(self,key):
    hash = 0
    m = self.MAX
    for c in key:
      hash += ord(c)
    return hash % m

  def __setitem__(self, key, value):
    h = self.hash_function(key)
    found = False


    for index, element in enumerate(self.arr[h]):
      if len(element) == 2 and element[0] == key:
        self.arr[h][index] = (key, value)
        found = True
        break

    if not found:
      self.arr[h].append((key,value))

  def __getitem__(self, key):
    h = self.hash_function(key)
    for index, element in enumerate(self.arr[h]):
      if len(element) == 2 and element[0] == key:
        return element[1]

  def __delitem__(self, key):
    h = self.hash_function(key)
    for index, element in enumerate(self.arr[h]):
      if element[0] == key:
        del self.arr[h][index]
```

```python
obj3 = HashTable()
obj3["march 6"] = 23
obj3["march 17"] = 27
```

```python
print(obj3.arr)
```

```python
obj3["march 6"]
```

```python
del obj3["march 6"]
```

```python
print(obj3.arr)
```
