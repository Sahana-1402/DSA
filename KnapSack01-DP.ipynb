Implementing 0/1 Knapsack problem using Dynamic Programming

Constraints - Max weight is given

Optimisation - To get the max profit

Using just recursion

```python
# Method implementation
def knapSack01(val, wt, W, N):
  # base case
  if N == 0 or W == 0:
    return 0
  # skip object
  if wt[N-1] > W:
    return knapSack01(val, wt, W, N-1)
  else:
    return max(knapSack01(val, wt, W, N-1),(val[N-1]+knapSack01(val, wt, W-wt[N-1], N-1)))


# Driver code
val = [1, 2, 3]
wt = [4, 5, 1]
W = 4
N = len(val)

print(knapSack01(val, wt, W, N))
```

Using memoization

```python
# method defination
def knapSack01(val, wt, W, N):
  table = [[None]*(W+1) for i in range(N+1)]
  if N == 0 or W == 0:
    return 0
  if table[N][W] is not None:
    return table[N][W]
  elif wt[N-1] > W:
    table[N][W] = knapSack01(val, wt, W, N-1)
    return table[N][W]
  else:
    table[N][W] = max(knapSack01(val, wt, W, N-1), val[N-1]+knapSack01(val, wt, W-wt[N-1], N-1))
    return table[N][W]

  return table[N][W]


# Driver code
val = [1, 2, 3]
wt = [4, 5, 1]
W = 4
N = len(val)

print(knapSack01(val, wt, W, N))
```

Using Tabulation

```python
# Method implementation
def knapSack01(val, wt, W, N):
  table = [[None]*(W+1) for i in range(N+1)]

  for i in range(N+1):
    for j in range(W+1):
      if i == 0 or j == 0:
        table[i][j] = 0
      elif wt[i-1] <= j:
        table[i][j] = max(table[i-1][j], val[i-1]+table[i-1][j-wt[i-1]])

      else:
        table[i][j] = table[i-1][j]

  print(table)
  return table[N][W]

# Driver code
val = [1, 2, 3]
wt = [4, 5, 1]
W = 4
N = len(val)

print(knapSack01(val, wt, W, N))
```
