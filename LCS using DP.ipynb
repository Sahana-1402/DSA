Longest Common Subsequence using Recursion

```python
s1 = 'bd'
s2 = 'abcd'

# method defination
def LCS(s1, s2, i, j):
  if i >= len(s1) or j >= len(s2):
    return 0
  elif s1[i] == s2[j]:
    return 1 + LCS(s1, s2, i+1, j+1)
  else:
    return max(LCS(s1, s2, i+1, j), LCS(s1, s2, i, j+1))

print(LCS(s1, s2, 0, 0))
```

LCS using memoization

```python
s1 = 'bd'
s2 = 'abcd'

# method defination
def getMemoizedLCS(s1, s2, i, j, matrix):
  if i >= len(s1) or j >= len(s2):
    return 0
  if matrix[i][j] != None:
    return matrix[i][j]
  elif s1[i] == s2[j]:
    result = 1 + getMemoizedLCS(s1, s2, i+1, j+1, matrix)
    matrix[i][j] = result
    return result
  else:
    result = max(getMemoizedLCS(s1, s2, i+1, j, matrix), getMemoizedLCS(s1, s2, i, j+1, matrix))
    matrix[i][j] = result
    return result


def LCS(s1, s2):
  matrix = [[None]*len(s2) for i in range(len(s1))]
  return getMemoizedLCS(s1, s2, 0, 0, matrix)

print(LCS(s1, s2))
```
